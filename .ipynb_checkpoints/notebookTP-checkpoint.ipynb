```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tabla1 = pd.read_csv("./output.csv", index_col="Tamaño Conjunto")
tabla1
```

```python
#list(tabla1)
listaPlot = [i for i in range(10, 29)]
graf1 = tabla1[["Fuerza Bruta", "Backtracking"]].plot(xticks=listaPlot,figsize=(10, 5), grid=True, title="Tiempo de ejecución entre los algoritmos de Fuerza Bruta y Backtracking")
graf1.set_xlabel("Tamaño del Conjunto")
graf1.set_ylabel("Tiempo de Ejecucion en milisegundos")
plt.savefig("G1.pdf")
```

```python
tabla2 = pd.read_csv("./output.csv", index_col="Tamaño Conjunto")
tabla2
```
